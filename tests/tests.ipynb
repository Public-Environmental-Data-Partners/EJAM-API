## /data

```python
# Set up /data endpoint queries
queries = [
  {"buffer":0,"shape":json.dumps(houston_zips)}, # Zips in Houston
  {"buffer": 1, "shape": json.dumps(simple_shape), "geometries": True},
  {"buffer": 4, "sites": [{"lat":33, "lon":-112}]}, # A single set of coordinates
  {"buffer": 4, "sites": [{"lat":33, "lon":-112}, {"lat":34, "lon":-114}]}, # Two or more coordinates
  {"buffer": 0, "fips": ["482012301001","482012302002", "482012302003"], "scale":"blockgroup"}, # FIPS
  {"buffer": 0, "fips": "DE", "scale": "blockgroup"}, # One state, blockgroup level results
  {"buffer": 0, "fips": "DE", "scale": "county"}, # One state, county level results
  {"buffer": 0, "fips": ["DE", "RI"], "scale": "county"}, # Two states, county level results
]
```

```python
# Execute /data endpoint
import time

for q in queries:
  time.sleep(5)
  try:
    url = "http://localhost:55000/data" #"https://ejamapi-84652557241.us-central1.run.app/data"
    response = requests.post(url, json=q)
    # Load response
    df = pandas.DataFrame.from_dict(response.json())
    display(df)
  except:
    print("error with ", q)
```

### sitenumber validation (regression: PR #51)
Invalid `sitenumber` values must return **400** (they could otherwise leak into the "Site N" report-header label, e.g. "Site 1.5" / "Site Inf"); valid forms still work. These fail fast (validated before the analysis runs).

```python
# GET /report: invalid sitenumber -> 400; valid forms -> not 400
url = "http://localhost:55000/report" #"https://ejamapi-84652557241.us-central1.run.app/report"
base = {"lat": 33.45, "lon": -112.07, "buffer": 1}
invalid = ["1.5", "Inf", "-Inf", "NaN", "-2", "abc"]
for s in invalid:
  r = requests.get(url, params={**base, "sitenumber": s})
  assert r.status_code == 400, f"sitenumber={s}: expected 400, got {r.status_code}"
  print("400 OK for sitenumber =", s)
# valid forms must NOT be rejected by validation (any non-400 means they got past it;
# 'overall'/0 and omitted are the aggregate/default paths, 1 is a plain single-site row)
for s in ["0", "overall", "1", ""]:
  r = requests.get(url, params={**base, "sitenumber": s})
  assert r.status_code != 400, f"sitenumber={s!r}: unexpected 400: {r.text[:200]}"
  print(r.status_code, "(not 400) for sitenumber =", repr(s))
```

```python
# POST /report: same validation as GET
url = "http://localhost:55000/report"
r = requests.post(url, json={"sites": [{"lat": 33.45, "lon": -112.07}], "buffer": 1, "sitenumber": 1.5})
assert r.status_code == 400, f"expected 400, got {r.status_code}"
print("400 OK for POST sitenumber = 1.5")
```
